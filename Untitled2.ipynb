```python

#tkinter import karlena UI ke liye
from tkinter import *
import tkinter.messagebox
from tkinter import filedialog
import os
from datetime import datetime;
import cv2
from PIL import Image 
import pickle
import numpy as np
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import smtplib
import random
# from twilio.rest import Client
import time
import tkinter as tk


face_dict = {}
emails=[]
phones=[]
#creating instance of TK
root=Tk()

dataset = pd.DataFrame()

root.configure(background="white")

#root.geometry("300x300")
def show_entry_fields( t,s):
    print(s)
    t.destroy()
    cam = cv2.VideoCapture(0)
    detector=cv2.CascadeClassifier('haarcascade_frontalface_default.xml')

    Id=s
    sampleNum=0
    try:
        os.mkdir("dataset")
    except:
        pass
    try:
        os.mkdir("dataset"+"\\"+Id)
    except:
        pass
    while(True):
        ret, img = cam.read()
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        faces = detector.detectMultiScale(gray, 1.3, 5)
        for (x,y,w,h) in faces:
            cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)

            #incrementing sample number 
            sampleNum=sampleNum+1
            #saving the captured face in the dataset folder
            cv2.imwrite('dataset/'+Id+'/'+Id +'.'+ str(sampleNum) + ".jpg", gray[y:y+h,x:x+w])

            cv2.imshow('frame',img)
        #wait for 100 miliseconds 
        if cv2.waitKey(100) & 0xFF == ord('q'):
            break
        # break if the sample number is morethan 20
        elif sampleNum>50:
            break
    cam.release()
    cv2.destroyAllWindows()
    messagebox.showinfo("Attendence Manager","FACE enrolled successfully")


def function1():


    master = tk.Tk()
    tk.Label(master, 
             text="Enrollment Number").grid(row=0)


    e1 = tk.Entry(master)
    e2 = tk.Entry(master)

    e1.grid(row=0, column=1)
    tk.Button(master, 
              text='Next', command=(lambda: show_entry_fields( master,e1.get()))).grid(row=3, 
                                                           column=1, 
                                                           sticky=tk.W, 
                                                           pady=2)

    
def function2():
    
    recognizer = cv2.face.LBPHFaceRecognizer_create()
    recognizer.read('trainer\\trainer.yml')
    images, labels = get_images_and_labels('dataset\\shubham')
    #cv2.imshow('test',images[0])
    #cv2.waitKey(1)
    save_obj(face_dict,"trainer\\trained_faces")
    recognizer.train(images, np.array(labels))
    recognizer.save('trainer\\trainer.yml')
    cv2.destroyAllWindows()


def function5():    
    root.destroy()

def get_images_and_labels(path):
    global face_dict
    names=os.listdir("dataset")
    image_paths=[]
    new_path="dataset"
    for name in names:
        image_paths.append([os.path.join(new_path+"\\"+name, f) for f in os.listdir(new_path+"\\"+name)])
    print(image_paths)
 # images will contains face images
    images = []
 # labels will contains the label that is assigned to the image
    labels = []
    nbr = 0
    i = 0
    j = 0
    for name in names:
        for image_path in image_paths[j]:
             # Read the image and convert to grayscale
            image_pil = Image.open(image_path).convert('L')
             # Convert the image format into numpy array
            image = np.array(image_pil, 'uint8')
            face_dict[j] = name
            print(face_dict)
            #nbr=int(''.join(str(ord(c)) for c in nbr))
            print(image_path)
             # Detect the face in the image
            faceCascade = cv2.CascadeClassifier('face.xml')
            faces = faceCascade.detectMultiScale(image)
             # If face is detected, append the face to images and the label to labels
            for (x, y, w, h) in faces:
                images.append(image[y: y + h, x: x + w])
                labels.append(j)
                cv2.imshow("Adding faces to traning set...", image[y: y + h, x: x + w])
                cv2.waitKey(10)
            i = i+1
        j = j+1
    return images,labels

def save_obj(obj, name ):
    with open(name + '.pkl', 'wb') as f:
        pickle.dump(obj, f, pickle.HIGHEST_PROTOCOL)

#stting title for the window
root.title("AUTOMATIC ATTENDANCE MANAGEMENT USING FACE RECOGNITION")

#creating a text label
Label(root, text="FACE SAVING FOR ATTENDENCE",font=("times new roman",20),fg="white",bg="maroon",height=2).grid(row=0,rowspan=2,columnspan=2,sticky=N+E+W+S,padx=5,pady=5)

#creating first button
Button(root,text="ADD FACE with ROLLNUMBER",font=("times new roman",20),bg="#0D47A1",fg='white',command=function1).grid(row=3,columnspan=2,sticky=W+E+N+S,padx=5,pady=5)
#creating second button
Button(root,text="Train ALL Dataset",font=("times new roman",20),bg="#0D47A1",fg='white',command=function2).grid(row=4,columnspan=2,sticky=N+E+W+S,padx=5,pady=5)
Button(root,text="Exit",font=('times new roman',20),bg="maroon",fg="white",command=function5).grid(row=5,columnspan=2,sticky=N+E+W+S,padx=5,pady=5)


root.mainloop()
```
